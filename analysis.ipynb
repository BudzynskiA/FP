```python
def get_stats(vals):
    return {
        "mean": vals.mean(),
        "std": vals.std(),
        "V": vals.std() / vals.mean() * 100,
        "median": vals.median(),
        "min": vals.min(),
        "max": vals.max(),
        "q1": np.percentile(vals, 25),
        "q3": np.percentile(vals, 75),
        "q": np.subtract(*np.percentile(vals, [75, 25])) / 2,
        "Vq": (np.subtract(*np.percentile(vals, [75, 25])) / 2) / vals.median() * 100
    }
```

```python
df["COUNTRY_LOAD_PLACE"] = df["COD_LP"].str[:2]
df["COUNTRY_DELIVERY_PLACE"] = df["COD_DP"].str[:2]
df["Relation"] = df["COD_LP"].str[:2] + ' to ' + df["COD_DP"].str[:2]
```

```python
df["NETTO_EURO"] = df['EURO'] + df['OTHER_COSTS']
df["NETTO_EURO_FOR_KM"] = df['NETTO_EURO'] / df['KM']
df_relation = df.groupby(["Relation"], as_index=False).NETTO_EURO_FOR_KM.mean()
df_relation["Load Country"] = df_relation["Relation"].str[:2]
df_relation["Delivery Country"] = df_relation["Relation"].str[6:8]
del df_relation['Relation']
relation = df_relation
relation = relation.pivot("Load Country", "Delivery Country", "NETTO_EURO_FOR_KM")
sns.set (rc = {'figure.figsize':(22, 40)})
ax = sns.heatmap(relation, cmap="YlGnBu", square=True, annot=True, cbar_kws={"shrink":0.25, 'label': '[€/km]'},linewidths=0.5,linecolor="grey")
```

```python
df['RELATION_TYPE'] = np.where((df['COUNTRY_LOAD_PLACE'] == df['COUNTRY_DELIVERY_PLACE']),'domestic', 'international') #1
domestic = df['RELATION_TYPE'] == 'domestic' #2
df_international = df[~domestic] #3
```
